# Cleaning and Spliting the data in test and train


Removing the uncessary data or columns where NAN is there

```python
from sklearn.model_selection import StratifiedShuffleSplit
print("Before {}".format(len(data)))
dataAfter = data.dropna(subset=["reviews.rating"])
# Removes all NAN in reviews.rating
print("After {}".format(len(dataAfter)))
dataAfter["reviews.rating"] = dataAfter["reviews.rating"].astype(int)
```

Splitting the data

```python
split = StratifiedShuffleSplit(n_splits=5, test_size=0.2)
for train_index, test_index in split.split(dataAfter,
                                           dataAfter["reviews.rating"]): 
    strat_train = dataAfter.reindex(train_index)
    strat_test = dataAfter.reindex(test_index)
```

Checking if the split is stratified proportionately

```python
print("Train Data")
strat_train["reviews.rating"].value_counts()/len(strat_train)*100
```

```python
print("Test Data")
strat_test["reviews.rating"].value_counts()/len(strat_test)*100
```

# Data Exploration

```python
reviews = strat_train.copy()
reviews.head(2)
asins_count_ix = reviews["asins"].value_counts().index

fig = plt.figure(figsize=(16,10))
ax1 = plt.subplot(211)
ax2 = plt.subplot(212, sharex = ax1)
reviews["asins"].value_counts().plot(kind="bar", ax=ax1, title="ASIN Frequency",color=['C0', 'C1', 'C2', 'C3', 'C4'])
plt.subplot(2,1,2)
sns.pointplot(x="asins", y="reviews.rating", order=asins_count_ix, data=reviews)
plt.xticks(rotation=90)
plt.show()
```

Analysis after reviewing the information
<ul>
    <li>The most frequently reviewed products have their average review ratings in the 4.5 – 4.8 range, with little variance.</li>
    <li>Average review for the first 4 ASINs are rated between 4.5 – 4.8, which is considered good overall reviews.</li>
    <li>ASINs with lower frequencies have high variance hence they not significant for our analysis due to high variance.</li>
    <li>On the other hand, due to their lower frequencies for ASINs with lower frequencies, we suggest that this is a result of lower quality products.</li>
    <li>The last 4 ASINs have no variance due to their significantly lower frequencies, and although the review ratings are a perfect 5.0, but we should not consider the significance of these review ratings due to lower frequency.</li>
</ul>

# Correlations

```python
corr_matrix = reviews.corr()
corr_matrix 
```

```python
counts = reviews["asins"].value_counts().to_frame()
avg_rating = reviews.groupby("asins")["reviews.rating"].mean().to_frame()
table = counts.join(avg_rating)
plt.scatter("asins", "reviews.rating", data=table)
table.corr()
```

From our analysis in data exploration above between ASINs and reviews.rating, we discovered that there are many ASINs with low occurrence that have high variances, as a result we concluded that theses low occurrence ASINs are not significant in our analysis given the low sample size.
